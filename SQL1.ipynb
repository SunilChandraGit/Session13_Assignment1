```python
import numpy as np
import pandas as pd
from pandas import DataFrame, Series
import sqlite3 as db
from pandasql import sqldf

pysqldf = lambda q: sqldf(q, globals())

from pandasql import sqldf, load_meat, load_births
```

```python
sqladb = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data", header=None)
sqladb.columns=['age', 'workclass', 'fnlwgt', 'education', 'education_num', 'marital_status', 'occupation', 'relationship', 'race', 'sex', 'capital_gain', 'capital_loss', 'hours_per_week', 'native_country', '-']
pysqldf("select * from sqladb LIMIT 10")
```

```python
q="""select sex, AVG(hours_per_week) from sqladb where sex = ' Male' group by sex"""
pysqldf(q)
```

```python
q="""select education, COUNT(*) from sqladb group by education"""
pysqldf(q)
```

```python
q="""select occupation, COUNT(*) from sqladb group by occupation"""
pysqldf(q)
```

```python
q="""select relationship, COUNT(*) from sqladb group by relationship"""
pysqldf(q)
```

```python
q="""select count(*) from sqladb where education=' Masters' and workclass = ' Private' and marital_status != ' Never-married'"""
pysqldf(q)
```

```python
q="""select workclass, MAX(age), MIN(age), AVG(age) from sqladb group by workclass"""
pysqldf(q)
```
