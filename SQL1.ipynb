```python
import pandas as pd
import numpy as np

#create the given dataframe
df = pd.DataFrame({'From_To': ['LoNDon_paris', 'MAdrid_miLAN', 'londON_StockhOlm',
'Budapest_PaRis', 'Brussels_londOn'],
'FlightNumber': [10045, np.nan, 10065, np.nan, 10085],
'RecentDelays': [[23, 47], [], [24, 43, 87], [13, 45, 89, 36, 54], [67, 32]],
'Airline': ['KLM(!)', '<Air France> (12)', '(British Airways. )',
'12. Air France', '"Swiss Air"']})

df
```

```python
val=0

#iterate through dataframe and update flight number for null values
for index, dr in df.iterrows():
    if(pd.isna(dr['FlightNumber'])):
        dr['FlightNumber']=val+10
        df.loc[index, 'FlightNumber']=val+10
    else:
        val=dr['FlightNumber']

#change column datatype to int 
df['FlightNumber']=df['FlightNumber'].astype(int)

df
```

```python
#create separate dataframe for from and to
fr = pd.DataFrame(df['From_To'].str.split('_').tolist(), columns=['From', 'To']) 
fr
```

```python
#capitalise values of from and to
fr['From'] = fr['From'].str.title()
fr['To'] = fr['To'].str.title()
fr
```

```python
#delete From_To column from dataframe
del df['From_To']

#join temporary table to permanent table
df=df.join(fr)
df
```

```python
#get maximum number of delays
maxEle = df['RecentDelays'].apply(lambda x: len(x)).max()

#get all delays in a temp. dataframe with separate column
ps = pd.DataFrame(df['RecentDelays'].tolist(), columns=['delay_'+str(i+1) for i in range(maxEle)])

#join it with original dataframe
df.join(ps)
```
